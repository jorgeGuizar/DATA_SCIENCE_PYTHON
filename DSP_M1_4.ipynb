### loc,iloc,ix

There are three ways to select data from a data frame in Pandas: loc, iloc, and ix. 

#### loc
loc is $\textbf{primarily label based}$; when two arguments are used you use column headers and row index to select the data you want. loc can also take an integer as a row or column number.

### ix
By default, ix looks for a label. If ix doesn't find a label, it will use an integer. This means you can select data by using either column numbers and row numbers or column headers and row names using ix.

In Pandas version 0.20.0 and later, ix is deprecated.

```python
# isiing the open mode W to create a file with name.txt
# the method .write is ised to write lines
# if it is already in the file  will be overwritten
# If we use the write method successively, each time it’s called, it will write to the file.
with open("C:/Users/LENOVO/Documents/AGILE_JNT/python_practiceJG/ejemplo2.txt","w") as file2:
    file2.write("This is line A\n")
    file2.write("This is line B\n")
    
        
    
    
```

```python
# in this we can write ecah element in a list toa file
# at the endo f the loop the fille will be closed
Lines=["This is line 1\n","This is lines 2\n","This is line 3\n","this is line 4\n"]
with open("C:/Users/LENOVO/Documents/AGILE_JNT/python_practiceJG/ejemplo3.txt","w") as file2:
    for line in Lines:
        file2.write(line)
```

```python
# on the other hand I want to write ina file a list of values 
X=np.linspace(10,50,25)
with open("C:/Users/LENOVO/Documents/AGILE_JNT/python_practiceJG/ejemplo4.txt","w") as file2:
    for xp in X:
        file2.write("\n{}".format(xp))
        file2.write("\nSquare of {} is {}\n".format(xp,xp**2))
        
    #file2.write("This is line B\n")
```
